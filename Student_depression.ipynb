```python
x = df1.drop('Depression',axis =1)
y = df1['Depression']
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 42,test_size = 0.2,stratify=y)
```

```python
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred_train = log_reg.predict(x_train)
cnf_matrix = confusion_matrix(y_train,y_pred_train)
print("The confusion matrix :\n",cnf_matrix)
print("*"*30)

acc_score = accuracy_score(y_train,y_pred_train)
print("The accuracy score :",acc_score)

clf_rpt = classification_report(y_train,y_pred_train)
print("The Classififcation report \n",clf_rpt)
```

```python
y_pred = log_reg.predict(x_test)
cnf_matrix = confusion_matrix(y_test,y_pred)
print("The confusion matrix :\n",cnf_matrix)
print("*"*30)

acc_score = accuracy_score(y_test,y_pred)
print("The accuracy score :",acc_score)

clf_rpt = classification_report(y_test,y_pred)
print("The Classififcation report \n",clf_rpt)
```
